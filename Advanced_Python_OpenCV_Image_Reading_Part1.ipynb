```python
import cv2 as cv
```

#### **Reading Image File**

```python
# Reading image file:

img = cv.imread('/content/drive/MyDrive/images/T-M-2316-Editable-Balloons_ver_2-1.webp')
```

```python
#Show actual Image
display(Image.fromarray(img))
```

#### **Convert Image to Gray scale Version**

```python
#### **Convert Image to Gray scale Version**
gray_scale = cv.cvtColor(img, cv.COLOR_BGR2GRAY)
display(Image.fromarray(gray_scale ))
```

```python
# Blur Grey scale image
inv_gray_scale = 255 - gray_scale
display(Image.fromarray(inv_gray_scale ))
```

```python
# Gaussian Blur grey scale image
blur_image =cv.GaussianBlur(inv_gray_scale, (35,35), 0)
```

```python
display(Image.fromarray(blur_image))
```

#### Image Dilation and Eroding 

```python
canny = cv.Canny(gray_scale,100,255)
```

```python
# Show updated image
display(Image.fromarray(canny))
```

```python
#Eroding the image.
# The erode function is used and usful to reduce noise (random variation of colour and brightness))

erode = cv.erode(canny,(3,3),iterations=3)
```

```python
# Show updated image
display(Image.fromarray(erode))
```

```python
# Dialating the image
dilate = cv.dilate(canny,(3,3),iterations=3)
```

```python
# Show updated image, fills in missing pixels from the eroded image
display(Image.fromarray(dilate))
```

#### **Rotate Image**

```python
# Rotate image by 90 degrees

imgr = cv.rotate(img,rotateCode=0)
display(Image.fromarray(imgr))
```

```python
# Rotate image by 180 degrees which flips image upside down
imgr2 = cv.rotate(img,rotateCode=1)
display(Image.fromarray(imgr2))
```

```python
# Rotate image by 70 degrees
imgr3 = cv.rotate(img,rotateCode=2)
display(Image.fromarray(imgr3))
```

#### **Tilting Image 20 degrees**

```python
h,w,c =img.shape
```

```python
center = (h//2,w/2)
```

```python
rotation_matrix =cv.getRotationMatrix2D(center,20,1.0)
```

```python
final_rotated = cv.warpAffine(img,rotation_matrix,(w,h))
```

```python
display(Image.fromarray(final_rotated))
```

```python
# Changing scale of tilted image:
rotation_matrix2 =cv.getRotationMatrix2D(center,20,0.50)
```

```python
final_rotated2 = cv.warpAffine(img,rotation_matrix2,(w,h))
```

```python
display(Image.fromarray(final_rotated2))
```

```python
rotation_matrix2 =cv.getRotationMatrix2D(center,20,0.30)
```

```python
final_rotated2 = cv.warpAffine(img,rotation_matrix2,(w,h))
```

```python
display(Image.fromarray(final_rotated2))
```

```python
# Changing direction of rotation:
rotation_matrix2 =cv.getRotationMatrix2D(center,-20,0.30)
```

```python
final_rotated2 = cv.warpAffine(img,rotation_matrix2,(w,h))
```

```python
display(Image.fromarray(final_rotated2))
```

#### **Flipping the Image**

```python
# Flip Image Vertically
imgvf = cv.flip(img, flipCode=0)
display(Image.fromarray(imgvf))
```

```python
# Flip Image Horizontally
imghf = cv.flip(img, flipCode=1)
display(Image.fromarray(imghf))
```

```python
# Flip Image Vertically and Horizontally
imgvhf = cv.flip(img, flipCode=-1)
display(Image.fromarray(imgvhf))
```

#### **Stack Images to display in single frame**

```python
import numpy as np
```

```python
hstack1 =np.hstack((img,imghf))
```

```python
hstack2 =np.hstack((imgvf,imgvhf))
```

```python
stacked_img =np.vstack((hstack1,hstack2))
```

```python
display(Image.fromarray(stacked_img))
```

```python
display(Image.fromarray(cropped))
```
